##### 💻 Código del cuaderno y retorno de valor 💻

Gracias a la función _**notebookutils.notebook.exit()**_, podemos devolver un valor de vuelta de forma controlada
- El valor devuelto siempre es de tipo string
- Puede almacenarse en una variable para poder usarlo posteriormente en el cuaderno principal o por otros recursos
- El método para la ejecución del cuaderno invocado una vez se ejecuta, el resto de celdas posteriores se ignoran

```python
print(f"El valor recibido es: {caller_input_value}")

# Devolvemos un valor de vuelta en función del parámetro recibido
if caller_input_value == 87:
    notebookutils.notebook.exit(f"Ejecutado correctamente el cuaderno notebook_single_call con valor {caller_input_value}")
else:
    notebookutils.notebook.exit(f"notebook_returned_value")
```
